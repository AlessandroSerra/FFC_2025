```python
import scipy.optimize as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def retta(x, m, q):
    return m*x + q
```

```python
x = np.linspace(0, 5, 20)
y = retta(x, 2, -1)
yerr = np.random.normal(loc=0, scale=0.5, size=20)
y_exp = y + yerr
```

```python
popt, pcov = sp.curve_fit(retta, x, y_exp)
```

```python
m_fit, q_fit = popt
y_fit = retta(x, m_fit, q_fit)
```

```python
plt.scatter(x, y_exp)
plt.plot(x, y_fit, color='red')
```

```python
dati = np.loadtxt('exp_g.txt', skiprows=2)
```

```python
# array[riga, colonna]
```

```python
tempi = dati[:, 0]
posizione = dati[:, 1]
```

```python
def f_g(x, G):
    return 0.5 * G * x**2
```

```python
popt, pcov = sp.curve_fit(f_g, tempi, posizione)
g_fit = popt[0]
```

```python
err = np.sqrt(np.diag(pcov))
```

```python
print(f'G vale {g_fit:.2f} +- {err[0]:.2f}')
```

```python
plt.scatter(tempi, posizione)
plt.plot(tempi, f_g(tempi, g_fit), color='red')
```
